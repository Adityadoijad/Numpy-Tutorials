```python
import numpy as np
```

### Linear Algebra

```python
a = np.ones((2,3))
print(a)
print()

b=np.full((3,2),2)
print(b)

np.matmul(a,b)
```

```python
#find the determinant
c = np.identity(3)
np.linalg.det(c)
```

```python
 #dot product
dot_product = np.dot(a, b)
# OR
dot_product = a @ b

print(dot_product)
```

```python
# transpose
a_T = a.T
print(a_T)
```

```python
# cross product
v1 = np.array([1, 2, 3])
v2 = np.array([4, 5, 6])

cross = np.cross(v1, v2)
print(cross)
```

```python
M = np.array([[4, 2],
              [1, 3]])

eigenvalues, eigenvectors = np.linalg.eig(M)

print("Eigenvalues:", eigenvalues,"\n")
print("Eigenvectors:\n", eigenvectors)
```

```python
#inverse 
inverse = np.linalg.inv(M)
print(inverse)
```

```python
#rank of matrix
rank = np.linalg.matrix_rank(M)
print(rank)
```

```python
# norm(magnitude)
norm = np.linalg.norm(v1)
print(norm)
```

Solving Linear Equation (IMP)

```python
A = np.array([[3,1],
              [1,2]])
B = np.array([9,8])

X = np.linalg.solve(A, B)
print(X)
```
